```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
march12 = pd.read_csv('OCC/OCC1_2_Counter_03_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
march34 = pd.read_csv('OCC/OCC3_4_Counter_03_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
april12 = pd.read_csv('OCC/OCC1_2_Counter_04_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
april34 = pd.read_csv('OCC/OCC3_4_Counter_04_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
may12   = pd.read_csv('OCC/OCC1_2_Counter_05_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
may34   = pd.read_csv('OCC/OCC3_4_Counter_05_2020_Filtered.gz', compression='gzip', header=None, sep=',', quotechar='"')
```

```python
frames = [march12,april12,may12]
result12 = pd.concat(frames)
result12.shape
```

```python
frames = [march34,april34,may34]
result34 = pd.concat(frames)
result34.shape
```

```python
data_options=[
    'Diameter Front-End.Gy-CCR-Initial.request',
    'Diameter Front-End.Gy-CCR-Update.request',
    'Diameter Front-End.Gy-CCR-Termination.request'
]
```

```python
result12=result12.loc[((result12[5].str.match(data_options[0])|result12[5].str.match(data_options[1]))|result12[5].str.match(data_options[2]))]
result12.drop([2,3,4], inplace=True , axis=1)
result12
```

```python
result34=result34.loc[((result34[5].str.match(data_options[0])|result34[5].str.match(data_options[1]))|result34[5].str.match(data_options[2]))]
result34.drop([2,3,4], inplace=True , axis=1)
result34
```

```python
uniqueValues =result12[0].unique()
df12 = pd.DataFrame(uniqueValues)
df12.set_index(0,inplace=True)

uniqueValues =result34[0].unique()
df34 = pd.DataFrame(uniqueValues)
df34.set_index(0,inplace=True)
df34
```

```python
copy_in=result12.loc[result12[1].str.match('occ1')&result12[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr=result12.loc[result12[1].str.match('occ1')&result12[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud=result12.loc[result12[1].str.match('occ1')&result12[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df12['occ1_in']=copy_in[6]
df12['occ1_tr']=copy_tr[6]
df12['occ1_ud']=copy_ud[6]

copy_in=result12.loc[result12[1].str.match('occ2')&result12[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr=result12.loc[result12[1].str.match('occ2')&result12[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud=result12.loc[result12[1].str.match('occ2')&result12[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df12['occ2_in']=copy_in[6]
df12['occ2_tr']=copy_tr[6]
df12['occ2_ud']=copy_ud[6]
df12
```

```python
copy_in=result34.loc[result34[1].str.match('occ3')&result34[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr=result34.loc[result34[1].str.match('occ3')&result34[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud=result34.loc[result34[1].str.match('occ3')&result34[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df34['occ3_in']=copy_in[6]
df34['occ3_tr']=copy_tr[6]
df34['occ3_ud']=copy_ud[6]

copy_in=result34.loc[result34[1].str.match('occ4')&result34[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr=result34.loc[result34[1].str.match('occ4')&result34[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud=result34.loc[result34[1].str.match('occ4')&result34[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df34['occ4_in']=copy_in[6]
df34['occ4_tr']=copy_tr[6]
df34['occ4_ud']=copy_ud[6]
df34
```

```python
df12['occ1_in']=df12['occ1_in'].diff()
df12['occ1_tr']=df12['occ1_tr'].diff()
df12['occ1_ud']=df12['occ1_ud'].diff()
df12['occ2_in']=df12['occ2_in'].diff()
df12['occ2_tr']=df12['occ2_tr'].diff()
df12['occ2_ud']=df12['occ2_ud'].diff()
df12
```

```python
df34['occ3_in']=df34['occ3_in'].diff()
df34['occ3_tr']=df34['occ3_tr'].diff()
df34['occ3_ud']=df34['occ3_ud'].diff()
df34['occ4_in']=df34['occ4_in'].diff()
df34['occ4_tr']=df34['occ4_tr'].diff()
df34['occ4_ud']=df34['occ4_ud'].diff()

df34
```

```python
df12['Traffic']=df12['occ1_in']+df12['occ1_tr']+df12['occ1_ud']+df12['occ2_in']+df12['occ2_tr']+df12['occ2_ud']
df12['Traffic'][0]=0
df12.reset_index(drop=True, inplace=True)
uniqueValues =result12[0].unique()
df12['Date']=uniqueValues
df12[['Date','Time']] = df12.Date.str.split(expand=True)
df12
```

```python
df34['Traffic']=df34['occ3_in']+df34['occ3_tr']+df34['occ3_ud']+df34['occ4_in']+df34['occ4_tr']+df34['occ4_ud']
df34['Traffic'][0]=0
df34.reset_index(drop=True, inplace=True)
uniqueValues =result34[0].unique()
df34['Date']=uniqueValues
df34[['Date','Time']] = df34.Date.str.split(expand=True)
df34
```

```python
uniqueValues =df34['Date'].unique()
data = pd.DataFrame(uniqueValues)

def traffic(a):
    return (df12.loc[df12['Date']==a]['Traffic'].sum()+df34.loc[df34['Date']==a]['Traffic'].sum())
  

data['Count']=data.apply(lambda x: traffic(x[0]),axis=1)


data
```

```python

plt.figure(figsize=(14,6))
plt.title("Line Chart for OCC  Traffic")
sns.lineplot(data=data['Count'], label="Voice Traffic")
plt.xlabel("Date")
```
