```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
roam = pd.read_excel('ROAMING_DATA_FEB_MAR_APR_MAY2020.xlsx')
```

```python
types=[
    'GPRS',
    'Mobile Terminating Call',
    'Mobile originated Call',
    'Short Message Originating Call'
]
```

```python
gprs=roam.loc[roam['CALL TYPE DESCRIPTION'].str.match(types[0])]
gprs.drop(['CALLTYPE', 'CALL TYPE DESCRIPTION','CDR COUNT'], inplace=True , axis=1)
mtc=roam.loc[roam['CALL TYPE DESCRIPTION'].str.match(types[1])]
mtc.drop(['CALLTYPE', 'CALL TYPE DESCRIPTION','CDR COUNT'], inplace=True , axis=1)
moc=roam.loc[roam['CALL TYPE DESCRIPTION'].str.match(types[2])]
moc.drop(['CALLTYPE', 'CALL TYPE DESCRIPTION','CDR COUNT'], inplace=True , axis=1)
smoc=roam.loc[roam['CALL TYPE DESCRIPTION'].str.match(types[3])]
smoc.drop(['CALLTYPE', 'CALL TYPE DESCRIPTION','CDR COUNT'], inplace=True , axis=1)
smoc
```

```python
gprs.plot.bar(x=0, y="VOLUME", rot=70, title="GPRS",figsize=(30,6))
plt.xlabel("Date")
```
